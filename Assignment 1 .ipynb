### **1. Before analyzing the experiment’s results, we want to verify that the experiment properly randomized users. To do this, we compare the treatment and control groups by the users’ characteristics.**

####**1ai. (10 pts) Use the *difference-in-means* estimator to compare the *proportions* of UK and US consumers in Treatment versus Control groups. (Hint: create two indicator variables using the “country” variable: “IsUK” and “IsUS”, and compute the difference for each variable separately).**


```python
## 1a) Difference-in-means estimator to compare the proportions of UK and US consumers in Treatment versus Control groups

# UK treatment & control
uk_treatment = df[(df['country'] == 'UK') & (df['treatment'] == 1)]
uk_control = df[(df['country'] == 'UK') & (df['treatment'] == 0)]

# US treatment & control
us_treatment = df[(df['country'] == 'US') & (df['treatment'] == 1)]
us_control = df[(df['country'] == 'US') & (df['treatment'] == 0)]
```

```python
# 1ai) Proportion of UK & US treatment
total_treatment_count = df[df['treatment'] == 1].shape[0]

uk_treatment_count = uk_treatment.shape[0]
uk_treatment_prop = uk_treatment_count / total_treatment_count
uk_treatment_percent = uk_treatment_prop

us_treatment_count = us_treatment.shape[0]
us_treatment_prop = us_treatment_count / total_treatment_count
us_treatment_percent = us_treatment_prop

print("Proportion of UK treatment rows: {:.3f}".format(uk_treatment_percent))
print("Proportion of US treatment rows: {:.3f}".format(us_treatment_percent))

#Proportion of UK & US control
total_control_count = df[df['treatment'] == 0].shape[0]

uk_control_count = uk_control.shape[0]
uk_control_prop = uk_control_count / total_control_count
uk_control_percent = uk_control_prop

us_control_count = us_control.shape[0]
us_control_prop = us_control_count / total_control_count
us_control_percent = us_control_prop

print("Proportion of UK control rows: {:.3f}".format(uk_control_percent))
print("Proportion of US control rows: {:.3f}".format(us_control_percent))
```

```python
# 1ai) Difference in means
diff_in_proportion_uk = uk_treatment_prop - uk_control_prop
print("Diff in proportion for UK: {:.3}".format(diff_in_proportion_uk))
diff_in_proportion_us = us_treatment_prop - us_control_prop
print("Diff in proportion for US: {:.3}".format(diff_in_proportion_us))
```

####**1aii. (5 pts) What do you conclude about the validity of the randomization in terms of user location?**

```python
# 1aii)
import math
stdev_uk_treatment = math.sqrt(uk_treatment_prop * (1-uk_treatment_prop)/total_treatment_count) #would n be total treatment count? ITT (evverone in treatment)
print("STDEV for UK treatment: {:.3}".format(stdev_uk_treatment))

stdev_us_treatment = math.sqrt(us_treatment_prop * (1-us_treatment_prop)/total_treatment_count) #would n be total treatment count? ITT (evverone in treatment)
print("STDEV for US treatment: {:.3}".format(stdev_us_treatment))

stdev_uk_control = math.sqrt(uk_control_prop * (1-uk_control_prop)/total_control_count) #would n be total treatment count? ITT (evverone in treatment)
print("STDEV for UK control: {:.3}".format(stdev_uk_control))

stdev_us_control = math.sqrt(us_control_prop * (1-us_control_prop)/total_control_count) #would n be total treatment count? ITT (evverone in treatment)
print("STDEV for US control: {:.3}".format(stdev_us_control))

#std assuming consumers in 2 groups aren't related
std_uk_prop = (stdev_uk_treatment ** 2) + (stdev_uk_control **2)
std_uk_prop_sq = np.sqrt(std_uk_prop)

std_us_prop = (stdev_us_treatment ** 2) + (stdev_us_control **2)
std_us_prop_sq = np.sqrt(std_us_prop)

print("STDEV UK: {:.3}".format(std_uk_prop_sq))
print("STDEV US: {:.3}".format(std_us_prop_sq))
```

```python
# 1aii) CI for treatment

#CI for UK
ci_lower_uk = diff_in_proportion_uk - (1.96 * std_uk_prop_sq)
ci_lower_uk_percent = ci_lower_uk * 100
ci_upper_uk = diff_in_proportion_uk + (1.96 * std_uk_prop_sq)
ci_upper_uk_percent = ci_upper_uk * 100

print("CI Lower: {:.3f}%".format(ci_lower_uk_percent))
print("CI Upper: {:.3f}%".format(ci_upper_uk_percent))

#CI for US
ci_lower_us = diff_in_proportion_us - (1.96 * std_us_prop_sq)
ci_lower_us_percent = ci_lower_us * 100
ci_upper_us = diff_in_proportion_us + (1.96 * std_us_prop_sq)
ci_upper_us_percent = ci_upper_us * 100

print("CI Lower: {:.3f}%".format(ci_lower_us_percent))
print("CI Upper: {:.3f}%".format(ci_upper_us_percent))
# CI includes 0, which means that there are no significant differences between the treatment and control group's proportion of US and UK (user location)
```

####**1bi. (5 pts) Verify the randomization by past sales. Use the difference-in-means estimator to compare the average sales in the 2 weeks before the experiment in the Treatment versus Control groups.** 

```python
## 1bi) Difference-in-means compare the average sales in the 2 weeks before the experiment in the Treatment versus Control groups.

# Treatment and control sales mean
treatment = df[df['treatment'] == 1]
mean_past_sales_treatment = treatment['past_sales'].mean()
print("Mean of past sales for treatment group: ${:.3f}".format(mean_past_sales_treatment))

control = df[df['treatment'] == 0]
mean_past_sales_control = control['past_sales'].mean()
print("Mean of past sales for control: ${:.3f}".format(mean_past_sales_control))

diff_past_sales = mean_past_sales_treatment - mean_past_sales_control
print("Diff of past sales between treatment & control: ${:.3f}".format(diff_past_sales))

#not much diff of past sales between treatment and control
```

```python
# 1bii) STDEV
std_treatment_ps = treatment['past_sales'].std() / np.sqrt(treatment.shape[0])
print("Treatment STDEV Presales:: ${:.3f}".format(std_treatment_ps))

std_control_ps = control['past_sales'].std() / np.sqrt(control.shape[0])
print("Control STDEV Presales: ${:.3f}".format(std_control_ps))

#std assuming consumers in 2 groups aren't related
std_presales = np.sqrt(std_treatment_ps**2 + std_control_ps**2)
print("STDEV Presales: ${:.3f}".format(std_presales))
```

```python
# 1bii) CI
ci_lower = diff_past_sales - (1.96 * std_presales)
ci_upper = diff_past_sales + (1.96 * std_presales)

print("CI Lower: ${:.3f}".format(ci_lower))
print("CI Upper: ${:.3f}".format(ci_upper))

# CI includes 0, which means that there are no significant differences between the treatment and control group's past sales??
```
