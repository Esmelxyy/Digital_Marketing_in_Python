```python
# Import packages
# Note: R users should use `pwr.t2n.test`. `pwr.t.test` only allows for 50-50 split
import numpy as np
from statsmodels.stats.power import tt_ind_solve_power as SolPower## Statistical power from TOT estimate:
np.round(SolPower(effect_size=(0.02/std_treatment_sales), # normalized difference between Group 1 & 2: min effect size / standard deviation
                  nobs1=(treatment_group_3),         # size of Group 1 (i.e. Treatment)
                  alpha=0.05,           # significance: prob. of Type I error
                  ratio=(control_group_3)/(treatment_group_3),            # ratio: Group 2 size/Group 1 size
                  alternative='two-sided')
, 6)  # round the results to 3 digits (adjust according to need)
```

#### 2c.

```python
# Define range of treatment assignments to try
treatment_assignments = np.arange(0.05, 0.96, 0.05)

# Calculate power for each treatment assignment
powers = []
for assignment in treatment_assignments:
    nobs1 = assignment * 500000
    nobs2 = 500000 - nobs1
    power = SolPower(effect_size=(0.02/std_treatment_sales),
                     nobs1=nobs1,
                     alpha=0.05,
                     ratio=nobs2/nobs1,
                     alternative='two-sided')
    powers.append(power)

# Find treatment assignment that maximizes power
max_power_idx = np.argmax(powers)
max_power_assignment = treatment_assignments[max_power_idx]

print(f"The maximum power ({np.max(powers):.4f}) is achieved with a treatment assignment of {max_power_assignment:.2%}")
```
